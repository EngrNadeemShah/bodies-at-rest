##### Import Libraries

```python
import os
import pickle
import numpy as np
import textwrap
import pandas as pd

print(os.getcwd())
```

### All Files & Directories (Names)

```python
for root, dirs, files in os.walk('.'):
    print(f'Root: {root}')
    print()

    print(f'Directories:')
    for i, dir in enumerate(dirs):
        print(f'Directory ({i+1:2d}/{len(dirs)}): {dir}')
    print()

    print(f'Files:')
    for i, file in enumerate(files):
        print(f'File ({i+1:2d}/{len(files)}): {file}')
    print('______________________________________________________')
```

### All Files -> All Keys (Shapes)

```python
for root, dirs, files in os.walk('.'):
    print(f'Root: {root}')
    print()

    for f_no, file in enumerate(files):
        if file.endswith('.p'):
            file_path = os.path.join(root, file)
            print(f'File ({f_no+1:2d}/{len(files)}): {file_path}')
            with open(file_path, 'rb') as f:
                data = pickle.load(f, encoding='latin1')
            
            for i, (k, v) in enumerate(data.items()):
                if k == 'images':
                    key_info = f'Key ({i+1:2d}/{len(data)}): {k}\t\t| {np.array(v).shape}'
                else:
                    key_info = f'Key ({i+1:2d}/{len(data)}): {k}\t| {np.array(v).shape}'
                print(textwrap.indent(key_info, '    '))

            print('______________________________________________________')
```

### Save to Spreadsheet __(.csv)__

```python
data_list = []

for root, dirs, files in os.walk('.'):
    for file in files:
        if file.endswith('.p'):
            file_path = os.path.join(root, file)
            with open(file_path, 'rb') as f:
                data = pickle.load(f, encoding='latin1')
            
            row = {'File': file_path}
            for k, v in data.items():
                row[k] = str(np.array(v).shape)
            data_list.append(row)

# Create a DataFrame
df = pd.DataFrame(data_list)

# # Save to CSV
# df.to_csv('synth_data.csv', index=False)

# Save to Excel
df.to_excel('synth_data.xlsx', index=False)
```
